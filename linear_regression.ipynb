```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
```

```python
# plot x: 'time' y: 'n'
df.plot(x='time', y='n', kind='line')
# plot x: 'time' y: 'n' with error bars
```

```python
# plot n column as histogram
df['n'].hist(bins=5, edgecolor='black', alpha=0.7)
import matplotlib.pyplot as plt
plt.xlabel('n')
plt.ylabel('Frequency')
plt.title('How many gorillas they see')
plt.grid(axis='y', alpha=0.75)
plt.show()
```
